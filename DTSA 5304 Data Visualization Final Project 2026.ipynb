```python
#Welcome to my Jupyter Notebook, here we are examing data on drug use by age and analyzing the data to find relationships
import pandas as pd
import altair as alt
data = pd.read_csv("drug-use-by-age.csv")
data.head()
```

```python
#lets examine the relationship between alcohol use and age
AgeAlcohol = alt.Chart(data).encode(
x = alt.X('age',title = 'Age (Years)',axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
y = alt.Y('alcohol_use', title = '% of the population that use alcohol',axis=alt.Axis(labelAngle=0, labelFontSize=14, titleFontSize=16)),
color = alt.Color('alcohol_frequency',title = 'Drinks in the past 12 months', scale = alt.Scale(scheme = 'reds')),
tooltip = ['age','alcohol_use','alcohol_frequency'])

AgeAlcohol.mark_bar().properties(width = 800, height = 300, title = {'text':'The Impacts of Age on Alcohol', 'fontSize':20}).interactive()
```

```python
#We will then apply the same analysis to find relationships between marijuana and age
AgeM = alt.Chart(data).encode(
x = alt.X('age',title = 'Age (Years)',axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
y = alt.Y('marijuana_use', title = '% of the population that use marijuana',axis=alt.Axis(labelAngle=0, labelFontSize=14, titleFontSize=16)),
color = alt.Color('marijuana_frequency',title = 'Marijuana use in the past 12 months', scale = alt.Scale(scheme = 'viridis')),
tooltip = ['age','marijuana_use','marijuana_frequency']).interactive()

AgeM.mark_bar().properties(width = 800, height = 300, title = {'text':'The Impacts of Age on Marijuana use', 'fontSize':20})
```

```python
#In the next three visualizations, we will see if there is a relationship between alcohol and other drugs (pain relievers, stimulats, and sedatives)
DataAlcoholvsPainRelivers= pd.read_csv("Alcohol_vs_pain_relievers.csv")
G4 = alt.Chart(DataAlcoholvsPainRelivers).encode(
x = alt.X('age',title = "Age",axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
y = alt.Y("pain_releiver_frequency",title = "Pain relievers used in the past year",axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
color = alt.Color("alcohol_use",title = "% of population that use alcohol",scale = alt.Scale(scheme = 'purplered')),
tooltip = ["alcohol_use","pain_releiver_frequency",'age']).interactive()

B = alt.Chart(DataAlcoholvsPainRelivers).encode(
x = alt.X('age',title = "Age",axis=alt.Axis(labelAngle=0, labelFontSize=14, titleFontSize=16)),
y = alt.Y("pain_releiver_frequency",title = "Pain relievers used in the past year",axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
)




A = G4.properties(width = 800, height = 300, title = {'text':'The Impact of Age and Alcohol on the use of Pain Relivers', 'fontSize':20})
A.mark_circle(size = 900) + B.mark_line(color = "black")
```

```python
AlcoholvsStimulants = pd.read_csv("Alcohol_vs_stimulants.csv")
G5 = alt.Chart(AlcoholvsStimulants).encode(
x = alt.X('age',title = "Age",axis=alt.Axis(labelAngle=0, labelFontSize=16, titleFontSize=16)),
y = alt.Y("stimulant_frequency",title = "Stimulants used in the past year",axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
color = alt.Color("alcohol_use",title = "% of population that use alcohol",scale = alt.Scale(scheme = 'redpurple')),
tooltip = ["alcohol_use","stimulant_frequency",'age']).interactive()


A = G5.properties(width = 800, height = 300, title = {'text':'How Age and Alcohol affects Stimulant Use', 'fontSize':20})
A.mark_bar(size = 30)
```

```python
AlcoholvsSedatives = pd.read_csv("Alcohol_vs_sedatives.csv")
G6 = alt.Chart(AlcoholvsSedatives).encode(
x = alt.X('age',title = "Age",axis=alt.Axis(labelAngle=0, labelFontSize=16, titleFontSize=16)),
y = alt.Y("sedative_frequency",title = "Frequency of Sedative Use",axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
color = alt.Color("alcohol_use",title = "% of population that use alcohol",scale = alt.Scale(scheme = 'reds')),
tooltip = ["alcohol_use","sedative_frequency",'age']).interactive()

B = alt.Chart(AlcoholvsSedatives).mark_line(color = "black").encode(x = 'age', y= "sedative_frequency")

A = G6.properties(width = 800, height = 300, title = {'text':'How Sedative Use is impacted by Age and Alcohol', 'fontSize':20})
A.mark_circle(size = 1800) + B
```

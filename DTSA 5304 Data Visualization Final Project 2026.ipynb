```python
#Welcome to my Jupyter Notebook, here we are examing data on drug use by age and analyzing the data to find relationships
import pandas as pd
import altair as alt
data = pd.read_csv("drug-use-by-age.csv")
data.head()
```

```python
#lets examine the relationship between alcohol use and age
AgeAlcohol = alt.Chart(data).encode(
x = alt.X('age',title = 'Age (Years)',axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
y = alt.Y('alcohol_use', title = 'Alcohol Use (%)',axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
color = alt.Color('alcohol_frequency',title = 'Drinks in the past 12 months', scale = alt.Scale(scheme = 'reds')),
tooltip = ['age','alcohol_use','alcohol_frequency'])

AgeAlcohol.mark_bar().properties(width = 800, height = 300, title = {'text':'The Impacts of Age on Alcohol', 'fontSize':16}).interactive()
```

```python
#We will then apply the same analysis to find relationships between marijuana and age
AgeM = alt.Chart(data).encode(
x = alt.X('age',title = 'Age (Years)',axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
y = alt.Y('marijuana_use', title = 'Marijuana Use (%)',axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
color = alt.Color('marijuana_frequency',title = 'Marijuana use in the past 12 months', scale = alt.Scale(scheme = 'viridis')),
tooltip = ['age','marijuana_use','marijuana_frequency']).interactive()

AgeM.mark_bar().properties(width = 800, height = 300, title = {'text':'The Impacts of Age on Marijuana use', 'fontSize':16})
```

```python
#In the next three visualizations, we will see if there is a relationship between alcohol and other drugs (pain relievers, stimulats, and sedatives)
G4 = alt.Chart(data).encode(
x = alt.X("alcohol_frequency",title = "Frequency of Alcohol Use",axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
y = alt.Y("pain_releiver_frequency",title = "Frequency of Pain Reliever Use",axis=alt.Axis(labelAngle=0, labelFontSize=12, titleFontSize=14)),
color = alt.Color('age',title = "Age",scale = alt.Scale(scheme = 'purplered')),
tooltip = ["alcohol_frequency","pain_releiver_frequency",'age']).interactive()

A = G4.properties(width = 800, height = 300, title = {'text':'The Impact of Alcohol on Pain Reliever Use', 'fontSize':16})
A.mark_circle(size = 900)
```
